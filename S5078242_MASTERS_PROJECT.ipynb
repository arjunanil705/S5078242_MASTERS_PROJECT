# STEP 3: RUNNING THE 3  ALGORITHMS ON THE DATASET:

This step involves taking the 3 algorithms which are Random Forest Regressor, Support Vector Regressor and Neural Networks and applying them on the dataset without any hyper parameter tuning 

```python
#RANDOM FOREST BASE ALGORITHM 
```

Initialziang the Random Forest Regressors with the base Hyper parameters:

```python
random_forest = RandomForestRegressor(max_depth=5,min_samples_split=2,min_samples_leaf=10,n_estimators=10,random_state=3)
```

```python
#APPLYING THE 5 FOLD CROSS VALIDATION TECHNIQUE TO GET THE ALGORITHM EFFICIENCY ON THE TRAINING SET
```

```python
#Validation function
n_folds = 5

def rmsle_cv(model):
    kf = KFold(n_folds, shuffle=True, random_state=42).get_n_splits(X_train_new)
    rmse= cross_val_score(model, X_train_new, Y_train, scoring="neg_mean_absolute_error", cv = kf)
    return(rmse)
```

```python
score = rmsle_cv(random_forest)
print("\nThe MAE and Standard Deviation for Random Forest Regressor on the Training data is: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
```

```python
t0= time.clock() #START TIME FOR TESTING DATA COMPUTATION
random_forest.fit(X_train_new, Y_train)
rforest_train_pred = random_forest.predict(X_test_new)
print("\nThe MAE for Random Forest Regressor on the Testing data is: {:.4f} \n".format(mean_absolute_error(Y_test, rforest_train_pred)))
t1 = time.clock() - t0 # END TIME FOR COMPUTATION 
print("Total Computational time for the Random Forest Regressor is: ", t1 ) # CPU seconds elapsed (floating point)
```

```python
#CHECKING FOR NORMALITY:
```

The next step is to run the Algorithm over 30 times since it is probabilistic in nature and to determine whether the results follow a Gaussian distribution or not.

```python
L_randomforest=[] #THE ARRAY TO STORE ALL THE RANDOM FOREST RESULTS OVER DIFFERENT SEEDS
```

```python
for i in range(0,30):
    model_rf = RandomForestRegressor(max_depth=5,min_samples_split=2,min_samples_leaf=10,n_estimators=10,random_state=i)
    model_rf.fit(X_train_new, Y_train)
    model_rf_test_pred = model_rf.predict(X_test_new)
    Q=abs(model_rf_test_pred-Y_test)
    M= mean_absolute_error(Y_test, model_rf_test_pred)
    print(i)
    print(M)
    L_randomforest.append(mean_absolute_error(Y_test, model_rf_test_pred))
```

```python
# Checking the normality of the Random Forest results:
stat, p = normaltest(L_randomforest)
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05 #Significance level set at 5%
if p > alpha:
	print('The Samples from the Random Forest Regressor follow a Gaussian distribution (fail to reject H0)')
else:
	print('The Samples from the Random Forest Regressor follow a non Gaussian distribution (reject H0)')
```

```python
#SUPPORT VECTOR REGRESSOR BASE ALGORITHM:
```

INITIATING THE SUPPORT VECTOR REGRESSOR OBJECT WITH THE BASE HYPER PARAMETERS:

```python
score = rmsle_cv(svr_linear)
print("\nThe MAE and Standard Deviation for Support Vector Regressor on the Training data is: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
```

```python
t2= time.clock() #START TIME OF TESTING DATA COMPUTATION FOR SUPPORT VECTOR REGRESSOR
svr_linear.fit(X_train_new, Y_train)
svr_linear_test_pred = svr_linear.predict(X_test_new)
print("\nThe MAE for Support Vector Regressor on the Testing data is: {:.4f} \n".format(mean_absolute_error(Y_test, svr_linear_test_pred)))
t3 = time.clock() - t2 # END TIME FOR COMPUTATION 
print("Total computational time for the Support Vector Regressor is: ", t3 ) # CPU seconds elapsed (floating point)
```

THE NEXT STEP IS TO RUN THE SUPPORT VECTOR REGRESSOR OVER 30 DIFFERENT SEEDS AND FIND WHETHER THE RESULTS FOLLOW A GAUSSIAN DISTRIBUTION OR NOT

```python
L_svr=[] #THE ARRAY TO STORE THE 3O RESULTS OF THE SUPPORT VECTOR REGRESSOR OVER DIFFERENT SEEDS
```

```python
for i in range(0,30):
    model_svr = LinearSVR(C=1.0,epsilon=0.1,tol=0.001,random_state=i)
    model_svr.fit(X_train_new, Y_train) 
    model_svr_test_pred = model_svr.predict(X_test_new)
    Q=abs(model_svr_test_pred-Y_test)
    M= mean_absolute_error(Y_test, model_svr_test_pred)
    print(i)
    print(M)
    L_svr.append(mean_absolute_error(Y_test, model_svr_test_pred))
```

```python
# compare samples
stat, p = normaltest(L_svr) #DOING THE NORMALITY CHECK
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05 #Significance level set at 5%
if p > alpha:
	print('The samples from the Support Vector Regressor follow a Gaussian distribution (fail to reject H0)')
else:
	print('The Samples from the Support Vector Regressor follow a non Gaussian distribution (reject H0)')
```

```python
#NEURAL NETWORKSBASE ALGORITHM
```

The first step is to configure the Keras environment so that reproducible results can be obtained in neural networks

```python
# Seed value (can actually be different for each attribution step)
seed_value= 0

# 1. Set `PYTHONHASHSEED` environment variable at a fixed value
import os
os.environ['PYTHONHASHSEED']=str(seed_value)

# 2. Set `python` built-in pseudo-random generator at a fixed value
import random
random.seed(seed_value)

# 3. Set `numpy` pseudo-random generator at a fixed value
import numpy as np
np.random.seed(seed_value)

# 4. Set `tensorflow` pseudo-random generator at a fixed value
import tensorflow as tf
tf.set_random_seed(seed_value)

# 5. Configure a new global `tensorflow` session
from keras import backend as K
session_conf = tf.ConfigProto(intra_op_parallelism_threads=1, inter_op_parallelism_threads=1)
sess = tf.Session(graph=tf.get_default_graph(), config=session_conf)
K.set_session(sess)
```

Initializing the Neural Networks with 2 Hidden Layers and 100 neurons in each hidden layer

```python

def build_regressor():
    regressor = Sequential()
    regressor.add(Dense(units=100,kernel_initializer='normal', input_dim=24,activation='relu'))#FIRST HIDDEN LAYER
    regressor.add(Dense(units=100,kernel_initializer='normal',activation='relu'))#THIRD HIDDEN LAYER
    regressor.add(Dense(units=1))#OUTPUT LAYER
    regressor.compile(optimizer='adam', loss='mean_absolute_error',  metrics=['mae'])# ASSIGNING THE LOSS METRIC OF MAE
    return regressor
```

```python

seed=3 #fixing a seed
np.random.seed(seed)
# evaluate model with Training dataset
estimator = KerasRegressor(build_fn=build_regressor, epochs=100, batch_size=50, verbose=3)
# evaluate using 5-fold cross validation
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=seed)
results = cross_val_score(estimator, X_train_new, Y_train, cv=kfold)
print("\nThe MAE for Neural Networks on the Training data is: {:.4f} \n".format(results.mean()))
```

```python
t4= time.clock() #START TIME OF TESTING DATA COMPUTATION FOR NEURAL NETWORK REGRESSOR
estimator.fit(X_train_new, Y_train)
estimator_test_pred = estimator.predict(X_test_new)
print("\nThe MAE for Neural Network Regressor on the Testing data is: {:.4f} \n".format(mean_absolute_error(Y_test, estimator_test_pred)))
t5 = time.clock() - t4 # END TIME FOR COMPUTATION OF NEURAL NETWORK REGRESSOR
print("Total Computational time for the Neural Network Regressor on the testing Data is: ", t5 ) # CPU seconds elapsed (floating point)
```

THE NEXT STEP IS TO RUN THE NEURAL NETWORK ALGORITHM OVER 30 SEEDS AND STORE THE RESULTS TO CHECK WHETHER THE RESULTS FOLLOW A GAUSSIAN DISTRIBUTION OR NOT
